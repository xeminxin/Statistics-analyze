```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# F 분포 모델링
dfn = 5         # 자유도 1
dfd = 10        # 자유도 2

# F 분포 시뮬레이션
num_simulations = 1000
f_values = np.random.f(dfn, dfd, num_simulations)
print(f_values)
```

```python
# F 분포 시뮬레이션 결과 시각화
plt.hist(f_values, bins=20, alpha=0.5)
plt.show()
```
