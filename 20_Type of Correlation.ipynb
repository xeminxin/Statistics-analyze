- 선형 관계 : 두 변수 간에 직선 형태로 관계가 있는 경우. 이 때 상관 계수를 구할 수 있다. 상관 계수는 -1에서 1 사이의 값을 가지며, 1에 가까울수록 강한 양의 상관관계, -1에 가까울 수록 강한 음의 상관관계, 0에 가까울수록 상관 관게가 없다는 것을 나타냄.

- 비선형 관계: 두 변수 간의 관계가 직선이 아닌 곡선 형태로 나타나는 경우. 선형관계와 달리 상관 계수로 측정하기 어려우며, 다른 방법으로 분석해야 한다. ex) 로지스틱 회귀, 다항 회귀

```python
# 선형 관계, 비선형 관계 시각화

# 데이터 생성
import numpy as np

# 선형 관계 데이터 생성
x_linear = np.linspace(0, 10, 100)
y_linear = 2 * x_linear + 1

# 비선형 관계 데이터 생성
x_nonlinear = np.linspace(-10, 10, 100)
y_nonlinear = x_nonlinear ** 2

# linspace() : 시작점과 끝점 사이를 지정한 개수로 일정하게 나누어 배열을 생성해줌
```

```python
import matplotlib.pyplot as plt

# 선형 관계 시각화
plt.scatter(x_linear, y_linear)
plt.title('Linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# 비선형 관계 시각화
plt.scatter(x_nonlinear, y_nonlinear)
plt.title('Nonlinear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## 상관 관계 해석 방법
- 산점도 그래프의 패턴 파악 : 산점도 그래프에서 데이터의 패턴을 파악하는 것도 상관관계를 해석하는데 중요하다. 선형적인 패턴이 보이는 경우 선형 상관관계가 있을 가능성이 높으며, 곡선이나 다른형태의 패턴이 보이는 경우 비선형 상관관계가 있을 가능성이 높다.
- 이상치의 영향 파악 : 이상치(outlier)가 있는 경우 상관분석 결과에 큰 영향을 미칠 수 있다. 따라서 이상치가 있는지 파악하고 이를 처리하는 것이 중요하다.
- 상관관계의 인과 관계 파악 : 상관관계는 인과관계를 나타내지는 않는다. 상관관계가 있어도 인과관계를 나타내지 않을 수 있으며, 상관관계가 없어도 인과관계가 있을 수 있다. 따라서 상관관계 결과를 인과관계로 해석하는 것은 조심해야 한다.

```python
# 상관 관계 간단한 실습
import seaborn as sns

# load tips dataset
tips = sns.load_dataset('tips')

# calculate the correlation matrix
corr = tips.corr()

# print the correlation matrix
print(corr)

# create a heatmap with the correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
