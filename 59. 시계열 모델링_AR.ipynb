# 시계열 모델링의 기본 가정
- 정상성 : 시계열 데이터의 통계적 특성이 시간에 따라 일정하게 유지되는 것
- 독립성 : 시계열 데이터는 시간 의존성을 가지지만, 일반적으로 이전 시점의 데이터가 현재 시점의 데이터에 영향을 주는 경우가 많다. 따라서 독립성 가정은 시계열 분석에서 일반적으로 성립되지 않는다.
- 잔차의 정규성 가정 : 시계열 모델링에서는 모델의 잔차가 정규분포를 따른다는 가정을 기반으로 한다. 잔차는 모델로부터 예측된 값과 실제 관측값 간의 차이를 의미

### 자기 상관성과 정상성
- 자기상관성 : 시계열 데이터 내에서 관측 값들 사이의 상관관계를 의미. 데이터가 시간에 따라 얼마나 유지되느지, 어떤 주기성을 가지는지 등을 파악하는 데 도움을 준다. -> 데이터 간의 시간적 관계를 파악하고 패턴을 이해하는 데 도움을 준다.
- 정상성 : 정상성을 가정하면 데이터의 평균, 분산 및 자기상관성이 시간에 따라 변하지 않는다. -> 데이터의 일관된 통계적 특성을 가정하여 모델링과 예측에 유용


### 이동평균 모델(MA)
시계열 데이터의 예측을 위해 사용되는 모델 중 하나. 현재 관측치를 이전 관측치들의 오차항(잔차)의 가중 평균으로 설명한다.
- 이동평균 : 이전 관측치들의 오차항의 가중 평균. 이동평균은 현재 시점에서의 예측값을 계산하는 데 사용된다.
- 오차항 : MA 모델에서는 시계열 데이터의 예측 오차를 이전 관측치들의 오차항으로 설명한다. 이전 관측치들의 오차항은 시계열 데이터의 불규칙한 패턴이나 예측할 수 없는 변동성을 나타내는 요소로 간주된다.
- 이동평균 차수 : MA 모델은 MA(q)로 표현되며, 'q'는 이동평균 차수를 나타낸다. 이동평균 차수는 몇 개의 이전 관측치들의 오차항을 사용하여 현재 시점의 예측값을 계산할지를 계산한다.
- 자기상관성 : MA 모델은 자기상관성이 없는 (무상관) 시계열 데이터를 가정한다. 즉 이전 관츠기들의 오차항들은 서로 상관되지 않는다고 가정<br>
ARIMA 모델의 구성 요소 중 하나이다.

### AR(자기회귀) 모델 및 MA(이동평균) 모델
- AR 모델 : 현재 시점의 값이 이전 p개의 시점의 값의 선형 조합으로 설명됨을 나타냄. 파라미터는 주로 최소자승법을 사용하여 추정된다.
- MA 모델 : 현재 시점의 값이 이전 q개의 오차항의 선형 조합으로 설명됨을 나타냄. 파라미터는 주로 최소자승법이나 최대 가능도 추정 등의 방법을 사용하여 추청된다.<br>
ARMA 모델이나 ARIMA 모델과 같은 다른 모델들과 함께 사용될 수 있으며, 이들 모델은 시계열 데이터 분석과 예측에 유용하다. 파라미터 추정은 통계적 방법이나 최적화 알고리즘을 사용하여 수행


### ARMA 모델과 ARIMA 모델
- ARMA 모델<br>
AR 모델과 MA 모델의 결합.<br>
시계열 데이터의 자기상관성과 이동평균성을 모두 고려하여 데이터를 설명하는 모델.
이전 p개의 시점의 값과 이전 q개의 오차항의 선형 조합으로 현재 시점의 값을 설명한다.<br>
AR 모델은 시계열 데이터의 추세를 설명 가능, 잡음 설명 가능, 자기상관성과 이동평균성을 동시에 고려하므로 다양한 시계열 데이터 패턴을 모델링 할 수 있다. 또, 백색 잡음 가정을 만족하는 데이터에 적합하다. 파라미터는 최소자승법, 최대 가능도 추정 등의 통계적 방법을 사용하여 추적.<br>
AR 모델, MA 모델, 차분 개념을 결합한 시계열 모델
- ARIMA 모델<br>
ARIMA 모델은 시계열 데이터의 자기상관성, 이동평균성, 추세 변동을 모두 고려하여 데이터를 설명하고 예측하는 데 사용<br>
해당 모델은 자기회귀와 이동평균 모델을 결합하면서도 추세 변동 반영 가능. 차분 개념을 포함해 데이터의 비정상성을 제거하고 정상성을 가진 데이터로 변환.<br>
이 모델은 정상성을 가진 데이터에 적합하다. 따라서 정상성이 없을 경우 차분 과정을 통해 정상성을 확보한 후 모델을 적용한다.<br>
ARIMA(p, d, q) 모델에서는 p는 AR 모델의 차수, d는 차분의 차수, q는 MA 모델의 차수를 의미<br>
파라미터는 최소자승법, 최대 가능도 추정 등의 통계적 방법을 사용하여 추정.<br>
또한, ACF와 PACF를 통해 최적의 차수를 결정할 수 있다.
- 최대 가능도 추정은 주어진 데이터에 대해 모델의 파라미터 값을 찾는 과정으로, 가능도 함수를 최대화 하는 파라미터 값을 추정한다.

#### ARMA 모델의 파라미터 추정
- 초기 추정치 설정 : ARMA 모델의 차수(p, q)를 결정하고 초기 파라미터 추정치를 설정한다.
- 최대 가능도 추정 : 초기 추정치를 기반으로 주어진 데이터의 로그 가능도 함수를 정의하고, 이를 최대화하는 파라미터 값을 찾는다. 이를 위해 수치적 최적화 알고리즘을 사용할 수 있다.

#### ARIMA 모델의 파라미터 추정
- 차분 : ARIMA 모델에서는 주어진 시계열 데이터가 정상성을 가지지 않을 수 있으므로, 먼저 데이터의 차분을 적용하여 정상성을 만든다.
- 초기 추정치 설정 : ARMA 모델의 차수 (p,q)와 차분(D)을 결저앟고 초기 파라미터 추정치를 설정한다.
- 최대 가능도 추정 : 초기 추정치를 기반으로 주어진 데이터의 로그 가능도 함수를 정의하고, 이를 최대화하는 파라미터 값을 찾는다. 이를 위해 수치적 최적화 알고리즘을 사용함<br>
파라미터 추정 과정에서는 주어진 데이터와 모델의 잔차(오차) 사이의 차이를 최소화 하는 방향으로 파라미터를 조정. 이를 통해 데이터에 가장 적합한 모델의 파라미터값을 찾아내고, 이를 기반으로 시계열 데이터의 예측을 수행함

### 모델 선택
모델 선택을 위한 정보 기준
- AIC : 일반적으로 통계모델의 적합도와 복잡성을 고려하여 모델을 평가하고 비교하는 데 사용됨. 로그-우도(log-likelihood) 값을 활용하여 계산됨. 로그-우도값은 모델이 주어진 데이터를 얼마나 잘 설명하는지를 나타내는 지표. AIC는 로그-우도 값에 모델의 복잡성을 페널티로 적용하여 계산됨<br>
AIC = -2 * 로그-우도 + 2 * k<br>
k는 모델의 파라미터 수를 의미, AIC는 로그-우도 값에 모델의 파라미터 수를 보상하는 패널티로 사용됨. 따라서 AIC가 낮을수록 모델이 더 우수하다고 판단, 동시에 복잡성을 최소화 하는 모델로 선택됨
- BIC : AIC와 유사한 정보기준. 다른 점은 모델의 복잡성에 페널티를 부여하여 과적합을 방지하는 데 더 초점을 둔다는 것
BIC = -2 * log-likelihood + k * log(n)<br>
n은 데이터의 샘플 수<br>
BIC 값이 가장 작은 모델을 선택하는 방식. 작은 BIC 값은 더 정확한 모델을 나타내며, BIC를 기준으로 모델 간 비교를 수행할 수 있음. 일반적으로 BIC 값이 작은 모델이 데이터를 잘 설명하면서도 파라미터 수를 효과적으로 제어한 모델로 간주됨
<br> 여러 모델을 생성하고 BIC 값을 계산하여 비교한 후, 가장 작은 BIC 값을 가지는 모델을 선택하면 된다.

### 모델 평가
잔차 분석 : 모델이 데이터에 얼마나 적합한지 평가하기 위해 사용. 모델이 예측한 값과 실제 관측값 사이의 차이이다.
- 잔차의 분포: 잔차는 정규분포를 따르는지 확인한다. 정규성을 확인하기 위해 잔차 플롯이나 정규성 검정을 사용 할 수 있다.
- 잔차의 독립성: 잔차가 독립적으로 분포 하는지 확인한다. 시계열 데이터의 경우, 잔차 간에 자기 상관이 없어야 한다.
- 잔차의 등분산성: 잔차의 분산이 독립변수나 예측값과 관련없이 일정한지 확인한다. 등분산성을 확인하기 위해 잔차플롯이나 등분산성검정을 사용할 수 있다.<br>
모델의 예측 능력 평가 : 모델이 실제 관측값을 얼마나 잘 예측하는지 평가
- 평균 제곱 오차 (MSE) : 작은 MSE 값은 예측의 정확성을 나타냄
- 평균 절대 오차 (MAE) : 작은 MAE 값은 예측의 정확성을 나타냄
- 결정 계수 (R-squared) : 0에서 1 사이의 값, 높을수록 모델의 설명력이 좋음

## AR 모델링

```python
import pandas as pd
import numpy as np

# 시계열 데이터 생성
np.random.seed(0)
n = 100 # 데이터 포인트 수
ar_params = [0.7] # 자기회귀 계수
ma_params = [] # 이동평균 계수
residuals = np.random.normal(0, 1, n) # 잔차 생성

# AR(1) 모델 생성
data = [0]  # 초기값 설정
for i in range(1, n) :
    data.append(ar_params[0] * data[i-1] + residuals[i])
    
# 데이터프레임 생성
df = pd.DataFrame({'Date' : pd.date_range(start='2022-01-01', periods=n, freq='D'), 'Value' : data})
df = df.set_index('Date')

# 데이터 확인
print(df.head())
```

```python
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.graphics.tsaplots import plot_acf

# AR(1) 모델 적합
model = AutoReg(df['Value'], lags = 1)
# AutoReg 클래스를 사용하여 AR(1) 모델을 정의, df['Value']는 시계열 데이터를 나타내는 열이며, lags=1은 1차 자기회귀 모델을 의미한다.
results = model.fit()

# AR(1) 모델 정보 출력
print(results.summary())

# ACF 그래프 출력
plot_acf(df['Value'], lags = 20)
# acf 그래프를 생성, lags=20은 그래프에서 표시될 시차(lag)의 수
```

const: 상수항(절편)에 대한 회귀계수. 해당 모델에서는 0.0363으로 추정 되었으며, 통계적으로 유의하지 않음을 나타내는 p-value (0.716)을 가지고 있다. 따라서, 이 모델에서 상수항은 통계적으로 유의미한 영향을 가지지 않는 것으로 해석할 수 있다.<br>
Value.L1: Lag 1에 대한 회귀계수다. 해당 모델에서는 0.7776으로 추정 되었으며, 통계적으로 매우 유의함을 나타내는 p-value (0.000)을 가지고 있다.<br>
따라서, 이전 시점의 값이 현재값에 유의미한 양의 영향을 가진다는 것을 나타낸다. 이는 AR(1) 모델이 이전 시점의 관측치를 사용하여 현재값을 예측하는 자기회귀모델임을 의미한다.
- coef(coefficient) : 모델의 계수 또는 회귀 계수를 나타냄.
- std err (standard error): 추정된 계수의 표준 오차를 나타낸다. 작을 수록 추정치의 신뢰성이 높다.
- z-score (z): 각 계수에 대한 z-score 값으로, 계수 추정치를 표준오차로 나눈 값. 계수의 유의성을 평가 하는데 사용됨. 절댓값이 클 수록 계수가 유의미한 것으로 판단 될 가능성이 높아짐
- P>|z| (p-value): z-score에 대응하는 p-value 값으로, 계수가 0일 때 대립가설이 기각 될 확률을 나타냄. 작을 수록 계수가 유의미 하다고 판단
- [0.025 0.975] (confidence interval): 추정된 계수의 신뢰 구간을 나타낸다. 95% 신뢰수준에서 계수가 존재 할 것으로 예상 되는 구간을 나타냄

#### ACF그래프
시계열 데이터의 자기상관 함수를 시각화 한 것
