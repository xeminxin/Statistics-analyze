```python
import numpy as np
from scipy import stats
```

```python
# 두 집단의 데이터를 생성, 랜덤하게 생성된 정규분포 데이터 사용
np.random.seed(42)

# 모수 검정용 데이터 생성
n = 100
mu1, mu2 = 0, 0.5
sigma = 1
data1 = np.random.normal(mu1, sigma, n)
data2 = np.random.normal(mu2, sigma, n)

print(data1, data2)
```

```python
# 비모수 검정용 데이터 생성
m = 30
data3 = np.random.uniform(0,1,m)
data4 = np.random.uniform(0.5, 1.5, m)

print(data3, data4)
```

```python
# 데이터 생성 완료 -> 모수 검정 실시 !!
t_stat, p_val = stats.ttest_ind(data1, data2)
print(t_stat, p_val)
```

```python
# 데이터 생성 완료 -> 비모수 검정 실시
u_stat, p_val = stats.mannwhitneyu(data3, data4)
print(u_stat, p_val)
```
