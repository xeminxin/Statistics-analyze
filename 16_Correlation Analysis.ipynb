## 상관 분석
두 변수 간의 관계를 살펴보기 위한 방법 중 하나<br>
두 변수가 서로 어떤 관련성을 가지고 있는지, 그리고 그 관련성이 어느 정도인지를 분석하는 방법<br>
상관 분석에서는 두 변수 간의 선형적인 관계를 분석한다. 따라서 두 변수 간의 관계가 비선형적일 경우 상관 분석을 적용하기 어렵다. 상관 분석 결과는 상관 계수라는 값으로 나타난다. 상관 계수는 -1부터 1까지의 값을 가지며, 0에 가까울수록 두 변수 간의 상관 관계가 약하다는 것을 나타낸다. 만약 상관 계수가 1에 가깝다면, 두 변수 간의 상관 관계가 강하다는 것을 나타낸다.

```python
# 상관 분석 간단한 실습

# seaborn 내장 데이터셋인 tips 데이터를 활용하여 식당에서 팁을 받은 금액과 식사 금액 사이의 상관 관계를 분석
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
print(tips)
```

```python
# total_bill과 tip 변수 간의 상관 계수 계산
corr = tips[['total_bill', 'tip']].corr()
print(corr)
```

```python
# seaborn의 regplot() 함수를 활용해 total_bill과 tip 변수 간의 산점도와 회귀선 시각화
sns.regplot(x='total_bill', y='tip', data=tips)
```

해당 그래프에서 x축은 total_bill 변수를 나타내며 y축은 tip 변수를 나타낸다. 각각의 점은 tips 데이터셋에서 한 행(row)을 나타내며, 각 점은 total_bill값과 tip 값의 쌍으로 구성된다. 또한, 해당 그래프에는 회귀선도 함께 그려져 있는데, 이 회귀선은 주어진 데이터를 가장 잘 설명할 수 있는 직선이다. 회귀선의 기울기는 상관계수를 나타내며, 이는 x 변수와 y 변수 사이의 선형적인 관계를 나타내는 값이다. 따라서 해당 그래프를 보면, total_bill값과 tip 값은 양의 상관 관계를 가지고 있으며, 이를 통해 total_bill값이 증가할수록 tip 값도 증가한다는 경향성을 확인 할 수 있다.
