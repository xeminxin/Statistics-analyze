## 교호작용 효과
두 개 이상의 독립 변수가 종속 변수에 영향을 미칠 때, 이들 독립 변수들 간의 상호작용이 종속 변수에 영향을 미치는 것
<br> 즉, 하나의 변수만 고려하는 것이 아니라 여러 변수 간의 조합이 종속 변수에 영향을 미치는 것
<br> 교호작용 효과는 종속 변수에 대한 모형의 설명력을 높일 수 있다. 또한 교호작용 효과를 고려하여 데이터를 분석하면, 독립 변수 간의 관계를 보다 정확하게 파악할 수 있다. 이는 예측 모델의 정확도를 높이는 데 도움이 된다.
<br> 일반적으로 상호작용 항을 추가해 분석한다.(두 독립 변수의 곱으로 나타낼 수 있음)

```python
# 성별과 객실 등급이 생존 여부에 미치는 영향 체크

import seaborn as sns
import matplotlib.pyplot as plt

# load titanic dataset
titanic = sns.load_dataset('titanic')

print(titanic)

# create a pointplot with Sex, Pclass, and Survived variables
sns.pointplot(x='sex', y='survived', hue='pclass', data=titanic)

# add legend
plt.legend(title='Pcalss', loc='best')

# set plot title
plt.title('Survival Rate by Sex and Pclass')
```

이 그래프를 통해, 여성이 남성보다 생존율이 높다는 것과, 1등급 객실에 탑승한 여성의 생존율이 가장 높다는 것을 확인할 수 있다. 또한, 3등급 객실에 탑승한 남성의 생존율이 가장 낮다는 것도 알 수 있다. 이처럼 교호작용 효과를 시각화 함으로써, 데이터의 패턴을 더욱 잘 파악할 수 있다.
