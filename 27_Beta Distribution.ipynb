```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
# 모수 설정
alpha = 2
beta = 5

# 랜덤으로 사용자 행동 데이터 생성 (0~1 사이의 값)
user_data = np.random.beta(alpha, beta, size=1000)
print(user_data)
```

```python
# 데이터 분포 시각화
plt.hist(user_data, bins=10, density=True, alpha=0.8)
plt.show()
```
