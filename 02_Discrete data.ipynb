```python
# 이산형 데이터 : 정수값으로 이루어진 데이터

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./data/college_data.csv')

# top10perc 컬럼은 대학교의 학생중 상위 10% 학생의 비율을 나타내는 이산형 데이터입니다.
# 이 컬럼을 사용하여 이산형 데이터의 시각화를 실습

print(data['top10perc'].unique())
```

```python
# top10perc 막대 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(data['private'], data['top10perc'])
plt.title('Bar Chart of Top 10% Scores by Private/Public')
plt.xlabel('Private/Public')
plt.ylabel('Top 10% Scores')
plt.show()
```

```python
# top10perc 원 그래프 그리기
plt.figure(figsize=(10, 6))
data.groupby('private')['top10perc'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pre Chart of Top 10% Scores by Private/Public')
plt.ylabel('')
plt.show()
```

```python
# top10perc 히스토그램 그리기
plt.figure(figsize=(10, 6))
plt.hist(data['top10perc'], bins=20)
plt.title('Histogram of Top 1-% Scores')
plt.xlabel('Top 1-% Scores')
plt.ylabel('Frequnecy')
plt.show()
```

```python
# 빈도수 간단한 실습
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
data = np.random.randint(1, 7, size=100)

# 빈도수 계산
unique, counts = np.unique(data, return_counts=True)

# 막대 그래프 시각화
fig, ax = plt.subplots()
ax.bar(unique, counts)
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Frequnecy Distribution')
plt.show()
```

```python
# 상대적 빈도수 간단한 시각화

import numpy as np
import matplotlib.pyplot as plt

"""
계산 방법
각 숫자 빈도수 체크
만약 예를 들어 5라는 숫자의 빈도수는 4번
4/11(데이터 총 수) = 0.363636
"""

# 예시 데이터 생성
data = np.array([1,2,2,3,3,4,4,5,5,5,5])
print(data)

# 데이터에서 각 값의 빈도수 계산
value_counts = np.bincount(data)
print(value_counts)

print(len(data))
relative_frequencies = value_counts / len(data)
```

```python
# 막대그래프로 상대빈도수 시각화

plt.bar(np.arange(len(value_counts)), relative_frequencies)
plt.xticks(np.arange(len(value_counts)), np.arange(len(value_counts))+1)
plt.xlabel('Value')
plt.ylabel('Relative  Frequency')
plt.title('Relative Frequency Bar Chart')
plt.show()
```

```python
# 백분율 간단한 시각화 실습

import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D', 'E']
percentages = [30, 20, 15, 10, 25]

fig, ax = plt.subplots()
ax.pie(percentages, labels=labels, autopct = '%1.1f%%')
ax.axis('equal')
ax.set_title('Percentages of Categories')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(labels, percentages)
ax.set_xlabel('Categories')
ax.set_ylabel('Percentages')
ax.set_title('Percentages of Categories')
plt.show()
```
