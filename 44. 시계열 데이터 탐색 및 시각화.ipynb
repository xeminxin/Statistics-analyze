## 시계열 데이터 탐색 및 시각화

#### 시계열 데이터의 구성 요소 : 추세
- 상향 추세
- 하향 추세
- 평평한 추세

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 랜덤 시계열 데이터 생성 총 100개
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods = 100, freq = 'D')
values = np.random.randn(100).cumsum()

# 데이터 프레임 생성
data = pd.DataFrame({'Date' : dates, 'Value' : values})

print(data)
```

```python
# 상향 추세 시각화
data['Upward_Trend'] = np.linspace(0, 50, len(data))   # 상향 추세 생성
plt.plot(data['Date'], data['Upward_Trend'], label='Upward Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Upward Trend')
plt.legend()
plt.show()
```

```python
# 하향 추세 시각화
data['Downward_Trend'] = np.linspace(50, 0, len(data))   # 하향 추세 생성
plt.plot(data['Date'], data['Downward_Trend'], label='Downward Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Downward Trend')
plt.legend()
plt.show()
```

```python
# 평평한 추세 시각화
data['Flat_Trend'] = np.full(len(data), 50)   # 평평한 추세 생성
# np.full : 데이터의 길이만큼 50이라는 값을 가지는 배열 생성

plt.plot(data['Date'], data['Flat_Trend'], label='Flat Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Flat Trend')
plt.legend()
plt.show()
```

### 시계열 데이터의 패턴 소개

* 추세 : 데이터가 일정한 방향으로 증가하거나 감소하는 경향을 보이는 경우. 장기적인 변동을 나타내며, 상향 추세, 하향 추세, 평평한 추세일 수 있다.
* 계절성 : 데이터에 주기적인 패턴이 존재하는 경우. 일정한 주기마다 반복되는 변동을 의미, 연간, 분기간, 월간, 주간 단위 등 다양한 주기를 가질 수 있다.
* 주기 : 데이터에 장기적인 주기성이 관찰되는 경우. 보다 긴 시간 단위에서 발생하는 변동을 의미하며, 경제적인 사이클이나 경기 변동과 관련될 수 있다.
* 불규칙성 : 데이터에 예측하기 어려운 불규칙한 변동이 존재하는 경우. 불규칙성은 예측 모델에 포함되지 않는 예외적인 사건이나 임시적인 변동을 나타낼 수 있음.


### 시계열 데이터의 변동성
데이터 값들 사이의 변동의 정도나 크기를 의미
* 분산 : 데이터 값들의 퍼짐도를 측정하는 지표. 평균값과의 거리의 제곱의 평균. 분산이 클수록 데이터 값들이 평균에서 멀리 퍼져있는 것을 의미
* 표준편차 : 분산의 양의 제곱근. 데이터 값들의 퍼짐 정도를 나타내는 지표. 표준편차가 크면 데이터 값들이 평균에서 상대적으로 더 멀리 퍼져 있는 것을 의미
* 변동계수 : 표준편차를 평균으로 나눈 값. 상대적인 변동의 정도를 나타내는 지표. 변동계수가 크면 상대적으로 큰 변동성을 가진 것으로 해석됨

변동성은 시계열 데이터의 불안정성을 파악하는 데 중요한 지표로 활용된다. 변동성이 크다면 데이터 값들이 불규칙하게 변동하며 예측이 어려울 수 있음. 변동성을 이해하고 적절히 처리하는 것은 시계열 분석과 예츠그 모델링에 있어서 중요한 요소이다.
