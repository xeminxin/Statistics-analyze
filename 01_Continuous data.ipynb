```python
# 데이터셋 불러오기
data = pd.read_csv("./data/BlackFriday.csv")
# print(data)

# User_ID별 구매 총 금액 계산
user_total_spent = data.groupby('User_ID')["Purchase"].sum()

print(user_total_spent)
```

```python
# 히스토그램 그리기
plt.hist(user_total_spent, bins=50)
plt.title("Histogram of Total Purchase Amount by User")
plt.xlabel('Total Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# 밀도 그래프 시각화
plt.figure()
user_total_spent.plot.kde()
plt.title('Density Plot of Total Purhase amount by User')
plt.xlabel('Total Purchase Amount')
plt.show()


# 구매 금액의 분포를 좀 더 부드럽게 표현
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
data = np.random.normal(loc=0, scale=1, size=10000)

# 중심 경향성 계산
maen = np.mean(data)
median = np.median(data)
mode = np.round(np.mean(data))

# 분포 시각화
fig, ax = plt.subplots()
ax.hist(data, bins=50)
ax.axvline(mean, color='red', linestyle='dashed', linewidth=2)
ax.axvline(median, color='green', linestyle='dashed', linewidth=2)
ax.axvline(mode, color='purple', linestyle='dashed', linewidth=2)
ax.legend(['Mean: {:.2f}'.format(mean), 'Median: {:.2f}'.format(median), 'Mode: {}'.format(mode)])
ax.set_xlabel('Data Values')
ax.set_ylabel('Frequency')
ax.set_title('Data Distribution')
plt.show()

# mean, median, mode값 구하고 시각화
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 두 변수 간의 상관관계 생성
np.random.seed(0)
x = np.random.normal(loc=0, scale=1, size=100)
y = x + np.random.normal(loc=0, scale=0.5, size=100)

# 산포도 그리기
fig, ax = plt.subplots()
ax.scatter(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')
plt.show()

# x와 y가 서로 양의 상관관계를 가짐
```

```python
# 이상치 간단한 시각화 - Box Plot

import numpy as np
import matplotlib.pyplot as plt


# 데이터 생성
data = np.concatenate([np.random.normal(0,1,900),
                      np.random.normal(8,2,100),
                      np.random.normal(-4, 2, 100)])

# 상자 그림 (Box Plot) 시각화
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xlabel('Data')
ax.set_ylabel('Values')
ax.set_title('Box Plot')
plt.show()


# median, 사분위, min, max, outlier(이상치) 여부 확인
# outlier : 상자 그림에서 박스 내부에 위치하지 않는 값(일반적인 범위를 벗어난 값)
# 상자 부분의 길이와 위치를 통해 데이터의 분산 정도 파악 가능
# 상자 길이가 짧고, 하단의 수염과 상단의 수염이 길게 나타나면 중앙값을 중심으로 분산되어 있고 outlier가 존재하지 않는 것

# 히스토그램(Histogram) 시각화
fig, ax = plt.subplots()
ax.hist(data)
ax.set_xlabel('Data')
ax.set_ylabel('Frequency')
ax.set_title('Histogram')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
data = np.random.randint(1, 7, size=100)

# 빈도수 계산
unique, counts = np.unique(data, return_counts=True)

# 막대그래프시각화
fig, ax = plt.subplots()
ax.bar(unique, counts)
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Frequency Distribution')
plt.show()
```

```python
# 상관관계 간단한 시각화

import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
np.random.seed(123)
x = np.random.normal(loc=0, scale=1, size=100)
y = 2*x + np.random.normal(loc=0, scale=1, size=100)

# 산점도 시각화
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')
plt.show()

# 상관관계 계산
corr = np.corrcoef(x, y)[0, 1]
print("상관계수:", corr)

# 상관계수의 값이 1에 가까울수록 두 변수간의 관계가 강해짐
# -1에 가까울수록 음의 상관관계가 강해짐
# 0에 가까울수록 두 변수간의 선형적인 관계가 약해지고, 관게가 없다
```
